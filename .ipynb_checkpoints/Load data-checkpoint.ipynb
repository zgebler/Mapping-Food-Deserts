### Problem Statement
 - What factors are the best predictors of an area being designated as a Food Desert?
  
  
### Methodology and Models
 - Classification of regions as Food Deserts using USDA metrics
 - Clustering of Counties to uncover relationships between regions
 
### Risks and Assumptions
 - Data is at the County Level for Food Environment Atlas (ENV), Census Tract for Food Access (ACC)
     - Will need to aggregate but ensure information is preserved
 - Want to avoid returning lists of correlations without any predictive power
 - Assumptions:
     - Enough data is captured about consumer behavior to uncover relationship between
         individuals and Govt. Spending
 
 - Data Sources:
     - https://www.ers.usda.gov/data-products/food-access-research-atlas/
     - https://www.ers.usda.gov/foodatlas/

```python
# Food Access Atlas
acc_df = pd.read_excel("./data/Access.xlsx", sheet_name=None)
acc_df.keys()
```

```python
census = acc_df['Food Access Research Atlas']
census.shape
```

```python
# Make Flag for Low-Access Tracts
census['la_flag'] = (census['LA1and10'] == 1) | (census['LA1and20'] == 1) | (census['LAhalfand10'] == 1) | (census['LATracts_half'] == 1) | (census['LATracts1'] == 1) | (census['LATracts10'] == 1 )| (census['LATracts20'] == 1) | (census['LATractsVehicle_20'] == 1)
census['la_flag'] = np.multiply(census['la_flag'], 1)
census['la_flag'].value_counts(normalize= True)
```

```python
# Create Food Desert Flag 
census['fd_flag'] = (census['LILATracts_1And10'] == 1) | (census['LILATracts_1And20'] == 1) | (census['LILATracts_Vehicle'] == 1) | (census['LILATracts_halfAnd10'] == 1) 
census['fd_flag'] = np.multiply(census['fd_flag'], 1)
census['fd_flag'].value_counts(normalize = True)
```
